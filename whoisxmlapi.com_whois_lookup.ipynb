# NOW... the function

```python
import pandas as pd
import numpy as np
import requests

def whoisxmlapai_lookup(key,domain):
    url="https://www.whoisxmlapi.com/whoisserver/WhoisService?apiKey="+key+"&domainName="+domain+"&outputFormat=JSON"
    print(url)
    r = requests.get(url).json()
    
    if 'ErrorMessage' in r:
        print("Problem with you key!!!")
    
    try:
        df = pd.DataFrame.from_dict(r['WhoisRecord'],orient='index').T
    except:
        df = pd.DataFrame({'domainName':domain}, index=[0],columns=['domainName','createdDate','updatedDate','expiresDate','registrarName','whoisServer','hostNames'])
    
    #For createdDate
    if 'createdDateNormalized' in df.columns:
        df['createdDate'] = df['createdDateNormalized']
    elif 'createdDate' in df.columns :
        df['createdDate'] = df['createdDate']
    elif 'registryData' in df.columns:
        if 'createdDate' in df['registryData'][0]:
            df['createdDate'] = df['registryData'][0]['createdDate']
        else:
            df['createdDate'] = np.nan
    else:
        df['createdDate'] = np.nan

    #For updatedDate
    if 'updatedDateNormalized' in df.columns:
        df['updatedDate'] = df['updatedDateNormalized']
    elif 'updatedDate' in df.columns :
        df['updatedDate'] = df['updatedDate']
    elif 'registryData' in df.columns:
        if 'updatedDate' in df['registryData'][0]:
            df['updatedDate'] = df['registryData'][0]['updatedDate']
        else:
            df['updatedDate'] = np.nan
    else:
        df['updatedDate'] = np.nan

    #For expiresDate
    if 'expiresDateNormalized' in df.columns:
        df['expiresDate'] = df['expiresDateNormalized']
    elif 'expiresDate' in df.columns :
        df['expiresDate'] = df['expiresDate']
    elif 'registryData' in df.columns:
        if 'expiresDate' in df['registryData'][0]:
            df['expiresDate'] = df['registryData'][0]['expiresDate']
        else:
            df['expiresDate'] = np.nan
    else:
        df['expiresDate'] = np.nan

    if 'registrarName' in df.columns:
        df['registrarName'] = df['registrarName']
    else:
        df['registrarName'] = np.nan

    if 'whoisServer' in df.columns:
        df['whoisServer'] = df['whoisServer']
    elif 'registryData' in df.columns:
        if 'whoisServer' in df['registryData'][0]:
            df['whoisServer'] = df['registryData'][0]['whoisServer']
        else:
            df['whoisServer'] = np.nan
    else:
        df['whoisServer'] = np.nan 

    if 'nameServers' in df.columns:
        if 'hostNames' in df['nameServers'][0]:
            df['hostNames']= " ".join(df['nameServers'][0]['hostNames'])
        else:
            df['hostNames'] = np.nan
    elif 'registryData' in df.columns:
        if 'nameServers' in df['registryData'][0]:
            if 'hostNames' in df['registryData'][0]['nameServers']:
                df['hostNames']= " ".join(df['registryData'][0]['nameServers']['hostNames'])
            else:
                df['hostNames'] = np.nan
        else:
                df['hostNames'] = np.nan                
    else:
        df['hostNames'] = np.nan
       
    return df[['domainName','createdDate','updatedDate','expiresDate','registrarName','whoisServer','hostNames']]
```
